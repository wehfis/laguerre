```python
%matplotlib widget
import matplotlib.pyplot as plt
```

```python
file = open('laguerre.csv',mode='r')

fig = plt.figure(figsize=(10, 5))

ax = fig.gca()
ax.axhline(color="grey", ls="--", zorder=-1)
ax.axvline(color="grey", ls="--", zorder=-1)
ax.grid(True)

for row in file:
    row_data = list(map(float, row[:-1].split(",")))
    ax.set_xlim(0, row_data[::2][-1])
    ax.plot(row_data[::2], row_data[1::2],alpha=0.8)
plt.show()
file.close()
```

```python
file = open('laguerre_sin.csv',mode='r')
data = file.read()
file.close()

d = list(map(float, data[:-1].split(",")))

fig = plt.figure(figsize=(10, 5))

ax1 = fig.gca()
ax1.axhline(color="grey", ls="--", zorder=-1)
ax1.axvline(color="grey", ls="--", zorder=-1)

xs = d[::2]
ys = d[1::2]
ax1.grid(True)
ax1.plot(xs, ys)
plt.show()
```

```python
file_laguerre_multiple = open('laguerre_multiple.csv',mode='r')
data_of_static_left_plots = file_laguerre_multiple.readline()
list_of_data_left = list(map(float, data_of_static_left_plots[:-1].split(",")))

fig, ax2 = plt.subplots(4, 2, sharex='col', sharey='row', figsize = (20, 20))

x_points_left = list_of_data_left[::2]
y_points_left = list_of_data_left[1::2]

file_laguerre_multiple.readline()
    
for i, row in enumerate(file_laguerre_multiple):
    ax2[i, 0].axhline(color="grey", ls = "--", zorder=-1)
    ax2[i, 0].axvline(color="grey", ls = "--", zorder=-1)
    ax2[i, 0].plot(x_points_left, y_points_left, label = 'sin')
    ax2[i, 0].set_xlabel(r'$x$', fontsize=14)
    ax2[i, 0].set_ylabel(r'$f(x)$', fontsize=14)
    ax2[i, 0].set_title(r'$f(t) = \sin(t - \pi/2) + 1$', fontsize=20)
    ax2[i, 0].legend(loc = 'lower center', fancybox=True, framealpha=0.9, shadow=True, borderpad=1)

    data_right = list(map(float, row[:-1].split(",")))
    ax2[i, 1].axhline(color="grey", ls = "--", zorder=-1)
    ax2[i, 1].axvline(color="grey", ls = "--", zorder=-1)
    ax2[i, 1].plot(data_right[::2], data_right[1::2], '-r', label = 'f')
    ax2[i, 1].set_xlabel(r'$x$', fontsize=14)
    ax2[i, 1].set_ylabel(r'$f(x)$', fontsize=14)
    ax2[i, 1].set_title(r'$\widetilde f^N(t)$', fontsize=20)
    ax2[i, 1].legend(loc = 'lower center', fancybox=True, framealpha=0.9, shadow=True, borderpad=1)
file_laguerre_multiple.close()
plt.show()
```
